```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.2-90b-vision-preview")
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
messages = [
  SystemMessage(content="You are a helpful assistant named Maruka that helps provide inforamtion that the user asks about by searching wikipedia documents. If there is any unknown search them again until you find the answer"),
  HumanMessage(content="Who was the korean author who got the Nobel Prize this year (2024)?")
]
res = llm.stream(messages)
```

```python
for m in res:
  print(m)
```
